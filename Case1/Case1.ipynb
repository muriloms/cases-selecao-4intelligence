```python
# Função para calcular o maior período de aumento e diminuição nos valores de TFP
def periodo_aumento_queda_valores_tfp(pais_analise):
    is_falling = False
    is_rising = False

    count_falling = 0
    count_rising = 0

    max_count_falling = 0
    max_count_rising = 0

    for i in range(1, len(pais_analise)):
        
        if pais_analise.rtfpna.iloc[i-1] - pais_analise.rtfpna.iloc[i] > 0:
            is_falling = True
            is_rising = False
            count_falling += 1

            if count_falling >= max_count_falling:
                max_count_falling = count_falling
                year_start_falling = pais_analise.year.iloc[i - max_count_falling]
                year_finish_falling = pais_analise.year.iloc[i]

        elif pais_analise.rtfpna.iloc[i-1] - pais_analise.rtfpna.iloc[i] <= 0:
            is_falling = False
            is_rising = True
            count_rising += 1

            if count_rising >= max_count_rising:
                max_count_rising = count_rising
                year_start_rising = pais_analise.year.iloc[i - max_count_rising]
                year_finish_rising = pais_analise.year.iloc[i]

        if is_falling == False:
            count_falling = 0

        if is_rising == False:
            count_rising = 0
            
    return max_count_falling, year_start_falling, year_finish_falling, max_count_rising, year_start_rising, year_finish_rising
```

```python
# Função para gerar relatório com as principais informações do banco de dados

def gerar_relatorio_analise(dataframe):
    '''
    Utilizar o Markdown para imprimir o resultado dessa função
    '''
    
    # Gerar variaveis para relatório

    # Período de análise
    per_analise = [min(dataframe.year), max(dataframe.year), len(dataframe.year.unique())]

    # Análise por país
    analise_por_pais = {}
    for country in dataframe.isocode.unique():

        min_value = dataframe[dataframe['isocode']==country].rtfpna.describe()[3]
        min_value_year = dataframe[(dataframe['isocode']==country) & (dataframe['rtfpna']==min_value)].year.unique()[0]
        max_value = dataframe[dataframe['isocode']==country].rtfpna.describe()[7]
        max_value_year = dataframe[(dataframe['isocode']==country) & (dataframe['rtfpna']==max_value)].year.unique()[0]
        mean = dataframe[dataframe['isocode']==country].rtfpna.describe()[1]
        std = dataframe[dataframe['isocode']==country].rtfpna.describe()[2]
        
        analise_periodo_up_down = periodo_aumento_queda_valores_tfp(df_tfp[df_tfp['isocode']==country])
        
        analise_por_pais.update({country:
                                        [
                                        np.round(min_value,4), np.round(min_value_year,4),
                                        np.round(max_value,4), np.round(max_value_year,4),
                                        np.round(mean,4), np.round(std,4),
                                        analise_periodo_up_down
                                        ]
                                })
        
    return ("## Relatório análise exploratória \n \
       \n **-- Período de análise --** \n \
       \n Ano inicial: {}  | Ano final: {} \n | Número total de anos: {} \n \
       \n **-- Análise por país --** \n \
       \n **-- USA** \n \
       \n Menor valor: {} em {} | Maior valor: {} em {} \n | Média: {} | Desvio-padrão: {}  \n \
       \n Maior período de queda: {}  ( {} a {} ) | Maior período de crescimento: {} ( {} a {} )  \n \
       \n **-- CAN** \n \
       \n Menor valor: {} em {} | Maior valor: {} em {} \n | Média: {} | Desvio-padrão: {}  \n \
       \n Maior período de queda: {}  ( {} a {} ) | Maior período de crescimento: {} ( {} a {} )  \n \
       \n **-- MEX** \n \
       \n Menor valor: {} em {} | Maior valor: {} em {} \n | Média: {} | Desvio-padrão: {}  \n \
       \n Maior período de queda: {}  ( {} a {} ) | Maior período de crescimento: {} ( {} a {} )  \n \
       ".format(
               per_analise[0],  per_analise[1],  per_analise[2],
               analise_por_pais['USA'][0], analise_por_pais['USA'][1], 
                analise_por_pais['USA'][2], analise_por_pais['USA'][3],
               analise_por_pais['USA'][4], analise_por_pais['USA'][5],
                analise_por_pais['USA'][6][0], analise_por_pais['USA'][6][1], analise_por_pais['USA'][6][2],
                analise_por_pais['USA'][6][3], analise_por_pais['USA'][6][4], analise_por_pais['USA'][6][5],
                analise_por_pais['CAN'][0], analise_por_pais['CAN'][1], 
                analise_por_pais['CAN'][2], analise_por_pais['CAN'][3],
               analise_por_pais['CAN'][4], analise_por_pais['CAN'][5],
                analise_por_pais['CAN'][6][0], analise_por_pais['CAN'][6][1], analise_por_pais['CAN'][6][2],
                analise_por_pais['CAN'][6][3], analise_por_pais['CAN'][6][4], analise_por_pais['CAN'][6][5],
                analise_por_pais['MEX'][0], analise_por_pais['MEX'][1], 
                analise_por_pais['MEX'][2], analise_por_pais['MEX'][3],
               analise_por_pais['MEX'][4], analise_por_pais['MEX'][5],
                analise_por_pais['MEX'][6][0], analise_por_pais['MEX'][6][1], analise_por_pais['MEX'][6][2],
                analise_por_pais['MEX'][6][3], analise_por_pais['MEX'][6][4], analise_por_pais['MEX'][6][5],

               )
      )
```

# Estacionariedade

## Teste Dickey-Fuller aumentado
O teste Dickey-Fuller aumentado é um tipo de teste estatístico chamado teste de raiz unitária.

A intuição por trás de um teste de raiz unitária é que ela determina com que intensidade uma série temporal é definida por uma tendência.

Existem vários testes de raiz unitária e o Dickey-Fuller aumentado pode ser um dos mais amplamente utilizados. Ele usa um modelo autoregressivo e otimiza um critério de informação entre vários valores diferentes de atraso.

A hipótese nula do teste é que as séries temporais podem ser representadas por uma raiz unitária, que não é estacionária (possui alguma estrutura dependente do tempo). A hipótese alternativa (rejeitando a hipótese nula) é que a série temporal é estacionária.

    Hipótese nula (H0): Se não for rejeitada, sugere que a série temporal tenha uma raiz unitária, o que significa que não é estacionária. Tem alguma estrutura dependente do tempo.
    Hipótese alternativa (H1): a hipótese nula é rejeitada; sugere que a série temporal não possui uma raiz unitária, o que significa que é estacionária. Não possui estrutura dependente do tempo.

Nós interpretamos esse resultado usando o valor-p do teste. Um valor p abaixo de um limite (como 5% ou 1%) sugere que rejeitemos a hipótese nula (estacionária); caso contrário, um valor p acima do limiar sugere que falhamos em rejeitar a hipótese nula (não estacionária).

    Valor p> 0,05: falha ao rejeitar a hipótese nula (H0), os dados têm uma raiz unitária e não são estacionários.
    Valor de p <= 0,05: Rejeita a hipótese nula (H0), os dados não têm uma raiz unitária e são estacionários.

```python
def test_stationarity(timeseries):
    
    #estatísticas contínuas
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()
    
    # Dickey-Fuller:
    print ('Teste Dickey-Fuller:')
    dftest = adfuller(timeseries.values, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
    
    #Plot
    plt.plot(timeseries, color='blue',label='Original')
    plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
```

```python
# Função para criar serie diferenciada
def difference(dataset, interval=1):
    diff = list()
    for i in range(interval, len(dataset)):
        value = dataset[i] - dataset[i - interval]
        diff.append(value)
    return pd.Series(diff)
```

```python
# Transformar dados em series temporais
ts_usa = df_tfp[df_tfp['isocode'] == 'USA'].rtfpna
ts_usa.index = df_tfp[df_tfp['isocode'] == 'USA'].year

ts_can = df_tfp[df_tfp['isocode'] == 'CAN'].rtfpna
ts_can.index = df_tfp[df_tfp['isocode'] == 'CAN'].year

ts_mex = df_tfp[df_tfp['isocode'] == 'MEX'].rtfpna
ts_mex.index = df_tfp[df_tfp['isocode'] == 'MEX'].year

# Log
ts_usa_log = np.log(ts_usa)
ts_can_log = np.log(ts_can)
ts_mex_log = np.log(ts_mex)

# Diferenciadas
ts_usa_diff = difference(ts_usa.values)
ts_can_diff = difference(ts_can.values)
ts_mex_diff = difference(ts_mex.values)
```

# Modelo ARIMA

```python
# carregar pacotes
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt

import warnings
warnings.filterwarnings("ignore")
```

```python
# Funcao para diferenciar serie
def difference(dataset, interval=1):
    diff = list()
    for i in range(interval, len(dataset)):
        value = dataset[i] - dataset[i - interval]
        diff.append(value)
    return np.array(diff)
```

```python
# Funcao para inverter valores da serie diferenciada
def inverse_difference(history, yhat, interval=1):
    return yhat + history[-interval]
```

```python
# Funcao para avaliar um modelo ARIMA para uma determinada ordem (p, d, q) e retornar RMSE
def evaluate_arima_model(X, arima_order):
    # prepare training dataset
    X = X.astype('float32')
    train_size = int(len(X) * 0.50)
    train, test = X[0:train_size], X[train_size:]
    history = [x for x in train]
    
    # make predictions
    predictions = list()
    for t in range(len(test)):
        # difference data
        diff = difference(history, 1)
        
        # Modelo ARIMA
        model = ARIMA(diff, order=arima_order)
        
        model_fit = model.fit(trend='nc', disp=0)
        
        yhat = model_fit.forecast()[0]
        yhat = inverse_difference(history, yhat, 1)
        
        predictions.append(yhat)

        history.append(test.iloc[t])
        
    # calculate out of sample error
    mse = mean_squared_error(test, predictions)
    rmse = sqrt(mse)
    return rmse
```

```python
# evaluate combinations of p, d and q values for an ARIMA model
def evaluate_models(dataset, p_values, d_values, q_values):
    dataset = dataset.astype('float32')
    best_score, best_cfg = float("inf"), None
    for p in p_values:
        for d in d_values:
            for q in q_values:
                order = (p,d,q)
                try:
                    mse = evaluate_arima_model(dataset, order)
                    if mse < best_score:
                        best_score, best_cfg = mse, order
                    print('ARIMA%s RMSE=%.3f' % (order,mse))
                except:
                    continue
    print('Best ARIMA%s RMSE=%.3f' % (best_cfg, best_score))
```

```python
# Definir parâmetros para encontrar melhor valor
p_values = range(0, 7)
d_values = range(0, 3)
q_values = range(0, 7)
```

```python
# Encontrar melhor modelo
print("USA")
evaluate_models(ts_usa, p_values, d_values, q_values)
```

```python
print("CAN")
evaluate_models(ts_can, p_values, d_values, q_values)
```

```python
print("MEX")
evaluate_models(ts_mex, p_values, d_values, q_values)
```
